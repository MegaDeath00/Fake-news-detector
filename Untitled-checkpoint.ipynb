```python
txt_clmn = 'text'
lbl_clmn = 'label'
```

```python
stopword = set(stopwords.words('english'))
stopword
```

```python
def preprocess_text(text):
    remove_puntuation = [c for c in text if c not in string.punctuation]
    new_txt = ''.join(remove_puntuation)
    remove_stopword = ([x for x in new_txt.split() if x.lower() not in stopword])
    return remove_stopword
```

```python
new_data = data[txt_clmn].apply(preprocess_text)
```

```python
new_data
```

```python
lemmatizer = WordNetLemmatizer()
def lemmatize_txt(text):
    lemmatized_txt = ' '.join([lemmatizer.lemmatize(i)for i in text])
    return lemmatized_txt
new_data1 = new_data.apply(lemmatize_txt)
new_data1
```

```python
vectorizer = TfidfVectorizer()

x = vectorizer.fit_transform(new_data1)
y = data[lbl_clmn]
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.3,random_state=42)
```

```python
def prediction_model(model,xtest,ytest):
    y_pred = model.predict(xtest)
    accuracy = accuracy_score(ytest,y_pred)
    con_mat = confusion_matrix(ytest,y_pred)
    confidence_lvl = model.predict_proba(xtest)[:,1]
    roc_area = roc_auc_score(ytest, confidence_lvl)
    fpr, tpr, thresholds = roc_curve(ytest, confidence_lvl)
    precision, recall, _ = precision_recall_curve(ytest, confidence_lvl)
    pr_area = auc(recall, precision) 
    plt.figure(figsize=(15,7))
    sns.heatmap(con_mat,annot=True, fmt='d', cmap='Purples',
            xticklabels=['Fake','Real'],
            yticklabels=['Fake','Real'])
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title('Confusion Matrix')
    plt.show()
    cls_rep = classification_report(ytest,y_pred)
    print(cls_rep)
    RocCurveDisplay.from_predictions(ytest, confidence_lvl)
    plt.Figure(figsize=(12,10))
    plt.plot([0,1],[0,1])
    plt.title('ROC CURVE')
    plt.grid(False)
    plt.show()
    return{
        'Accuracy' : accuracy,
        'Confusion Matrix': con_mat,
        'ROC AUC' : roc_area,
        'PR AUC' : pr_area,
        'FPR' : fpr,
        'TPR' : tpr,
        'PRECISION' : precision,
        'RECALL' : recall
        
    }
    
 
```

```python
M_naive_Bayes = MultinomialNB(alpha=1.0, fit_prior=True, class_prior=None)
M_naive_Bayes.fit(xtrain,ytrain)
Multinomial_model = prediction_model(M_naive_Bayes,xtest,ytest)
Multinomial_model
```

```python
B_naive_Bayes = BernoulliNB(alpha=1.0, binarize=0.0, fit_prior=True, class_prior=None)
B_naive_Bayes.fit(xtrain, ytrain)
Bernoulli_Model = prediction_model(B_naive_Bayes,xtest,ytest)
Bernoulli_Model
```

```python
lr = LogisticRegression(max_iter=1000)
lr.fit(xtrain, ytrain)
Logistic_regression_model = prediction_model(lr, xtest, ytest)
Logistic_regression_model
```

```python
def plot_histogram(value, model_name, value_label):
    fig,ax = plt.subplots(figsize = (15,8))
    Bar_graph = plt.bar(model_name,value)
    plt.xlabel('Prediction Models')
    plt.ylabel(value_label)
    plt.title(f'{value_label} of Prediction Models')

    for bars in Bar_graph:
        y = bars.get_height()
        ax.text(bars.get_x() + bars.get_width()/2,  y, round(y,4), ha='center', va = 'bottom', color= 'black', fontweight= 'bold')

    plt.show()

acc_value = [Multinomial_model['Accuracy'], Bernoulli_Model['Accuracy'],Logistic_regression_model['Accuracy']]
models = ['Muktinomial Model', 'Bernoulli Model', 'Logistic Regression Model']
plot_histogram(acc_value,models,'Accuracy')    
```
